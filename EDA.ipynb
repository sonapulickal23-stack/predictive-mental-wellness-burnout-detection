SUMMARY STATISTICS

```python
daily_activity.describe()
```

### Observation: Daily Activity Dataset

The Daily Activity dataset contains 457 records. The average daily step count is approximately 6,547 steps, while the median is 5,986 steps. The average calories burned is around 2,189 calories per day. Some records show 0 steps and 0 calories, which may need further investigation during preprocessing. Sedentary minutes are high, with an average of about 995 minutes per day.

```python
sleep_day.describe()
```

### Observation: Sleep Day Dataset

The Sleep Day dataset contains 413 sleep records. Participants slept for an average of approximately 419 minutes (about 7 hours) per night, while the average time spent in bed was around 459 minutes (about 7.6 hours). This indicates that participants spent approximately 40 minutes awake while in bed on average. Most participants had one sleep record per day, with a maximum of three sleep records recorded for a single day.

```python
heart_rate.describe()
```

### Observation: Heart Rate Dataset

The Heart Rate dataset contains 1,154,681 heart rate measurements collected at one-second intervals. The average heart rate is approximately 80 beats per minute (bpm), with a minimum value of 36 bpm and a maximum value of 185 bpm. The large number of observations makes this the largest dataset in the project and provides detailed physiological information for analysing user activity and health patterns.

```python
weight.describe()
```

### Observation: Weight Dataset

The Weight dataset contains 67 records of participants' body weight and BMI measurements. The average body weight is approximately 72.04 kg, with an average BMI of 25.19, which falls within the overweight category according to standard BMI classifications. The `Fat` column contains data for only two records, making it unsuitable for further analysis. Therefore, this column will be removed during the data preprocessing stage.

```python
sleep_lifestyle.describe()
```

### Observation: Sleep Lifestyle Dataset

The Sleep Lifestyle dataset contains 374 participant records. The average age of participants is approximately 42 years. On average, participants sleep 7.13 hours per night and report a sleep quality score of 7.31 out of 10. The average daily physical activity level is 59.17, while the average stress level is 5.39. Participants have an average resting heart rate of 70 beats per minute and take approximately 6,817 steps per day. This dataset provides valuable lifestyle and health information that can be used to analyse factors affecting sleep quality and overall well-being.
