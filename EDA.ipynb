### Observation: Daily Activity Dataset

The Daily Activity dataset contains 457 records. The average daily step count is approximately 6,547 steps, while the median is 5,986 steps. ### Observation: Sleep Day Dataset

The Sleep Day dataset contains 413 sleep records. Participants slept for an average of approximately 419 minutes (about 7 hours) per night, while the average time spent in bed was around 459 minutes (about 7.6 hours). Most participants had one sleep record per day, with a maximum of three sleep records recorded for a single day.

```python
heart_rate.describe()
```

### Observation: Heart Rate Dataset

The Heart Rate dataset contains 1,154,681 heart rate measurements collected at one-second intervals. The average heart rate is approximately 80 beats per minute (bpm), with a minimum value of 36 bpm and a maximum value of 185 bpm. The large number of observations makes this the largest dataset in the project and provides detailed physiological information for analysing user activity and health patterns.

```python
weight.describe()
```

### Observation: Weight Dataset

The Weight dataset contains 67 records of participants' body weight and BMI measurements. The average body weight is approximately 72.04 kg, with an average BMI of 25.19, which falls within the overweight category according to standard BMI classifications. The `Fat` column contains data for only two records, making it unsuitable for further analysis. Therefore, this column will be removed during the data preprocessing stage.

```python
sleep_lifestyle.describe()
```

### Observation: Sleep Lifestyle Dataset

The Sleep Lifestyle dataset contains 374 participant records. The average age of participants is approximately 42 years. On average, participants sleep 7.13 hours per night and report a sleep quality score of 7.31 out of 10. The average daily physical activity level is 59.17, while the average stress level is 5.39. Participants have an average resting heart rate of 70 beats per minute and take approximately 6,817 steps per day. This dataset provides valuable lifestyle and health information that can be used to analyse factors affecting sleep quality and overall well-being.

## .5 Correlation Heatmap

```python
plt.figure(figsize=(12,8))

correlation = daily_activity.corr(numeric_only=True)

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap of Daily Activity Features")

plt.tight_layout()
plt.show()
```

### Observation

The correlation heatmap illustrates the relationships between numerical variables in the Daily Activity dataset. Total Steps, Total Distance, and Tracker Distance exhibit very strong positive correlations (close to 1.0), indicating that these variables measure similar aspects of physical activity. Calories burned also show a moderate positive correlation with Total Steps and Very Active Minutes, suggesting that increased physical activity leads to higher energy expenditure. Sedentary Minutes display negative correlations with several activity-related variables, indicating that more sedentary behavior is associated with lower physical activity levels.

## .6 Boxplot for Total Steps

The boxplot summarizes the distribution of total daily steps and highlights potential outliers. Most participants recorded between approximately 2,000 and 10,000 steps per day, with a median of around 6,000 steps. ## .8 Sleep Disorder Distribution

```python
plt.figure(figsize=(8,5))

sleep_lifestyle["Sleep Disorder"].fillna("No Disorder").value_counts().plot(
    kind="bar",
    edgecolor="black"
)

plt.title("Distribution of Sleep Disorders")
plt.xlabel("Sleep Disorder")
plt.ylabel("Number of Participants")

plt.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
### Observation

The bar chart illustrates the distribution of sleep disorders among participants. Most individuals did not report any sleep disorder, while smaller groups were diagnosed with sleep apnea or insomnia. The dataset is therefore imbalanced, with the "No Disorder" class containing considerably more observations than the other two categories. This class distribution should be considered during machine learning model development and evaluation.
```
